```python

# Load data in chunks
chunks = pd.read_csv(r'C:\Users\Shobhit Chauhan\OneDrive\Desktop\POTATA task\correct_twitter_202102.tsv', 
                     sep='\t', 
                     chunksize=10000,  # Adjust chunk size as needed
                     low_memory=False)

# Process the chunks
df_list = []
for chunk in chunks:
    df_list.append(chunk)

# Concatenate all chunks into a single DataFrame
df = pd.concat(df_list, ignore_index=True)
```

```python
def tweets_by_day(df, term):
    return df[df['id'].str.contains(term, case=False)].groupby(df['created_at'].dt.date).size()
```

```python
def unique_users(df, id):
    return df[df['id'].str.contains(id, case=False)]['user-id'].nunique()
df
```

```python
def tweet_locations(df, from_quote_search):
    return df[df['id'].str.contains(from_quote_search, case=False)]['place_id'].value_counts()
```

```python
def times_of_day(df, term):
    return df[df['id'].str.contains(term, case=False)]['created_at'].dt.hour.value_counts()
```

```python
def top_user(df, term):
    return df[df['id'].str.contains(term, case=False)]['user_id'].value_counts().idxmax()
```
